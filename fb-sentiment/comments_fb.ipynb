```python
import json
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
if 'comments_v2' in comment_history:
    comments = comment_history['comments_v2']
    
    for comment_entry in comments:
        comment_data = comment_entry.get('data', [])
        for comment_item in comment_data:
            comment = comment_item.get('comment', {})
            comment_text = str(comment.get('comment', None))

            sentiment_scores = sia.polarity_scores(comment_text)
            compound_score = sentiment_scores['compound']

            print(f"Comment: {comment_text}")
            print(f"Sentiment Polarity: {compound_score}")
            print("----")

else:
    print("No 'comments_v2' key found in the JSON data.")
```

```python
from langdetect import detect
import re
import unicodedata
```

```python
num_comments = 0
num_english_comments = 0
num_non_english_comments = 0

if 'comments_v2' in comment_history:
    comments = comment_history['comments_v2']
    
    for comment_entry in comments:
        comment_data = comment_entry.get('data', [])
        for comment_item in comment_data:
            comment = comment_item.get('comment', {})
            comment_text = str(comment.get('comment', None)) 

            cleaned_comment = re.sub(r'[^a-zA-Z0-9\s]', '', comment_text)

            if not cleaned_comment.strip():
                continue

            num_comments += 1

            cleaned_comment_lower = cleaned_comment.lower()

            try:
                lang = detect(cleaned_comment_lower)
                if lang == 'en':
                    num_english_comments += 1
                    is_english = True
                else:
                    num_non_english_comments += 1
                    is_english = False
            except:
                is_english = False

            sentiment_scores = sia.polarity_scores(cleaned_comment_lower)
            compound_score = sentiment_scores['compound']

            print(f"Original Comment: {comment_text}")
            print(f"Cleaned Comment: {cleaned_comment}")
            print(f"Sentiment Polarity: {compound_score}")
            print(f"Language: {'English' if is_english else 'Not English'}")
            print("----")

print(f"Total Number of Comments: {num_comments}")
print(f"Number of English Comments: {num_english_comments}")
print(f"Number of Non-English Comments: {num_non_english_comments}")
```
